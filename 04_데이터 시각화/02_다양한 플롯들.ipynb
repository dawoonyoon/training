```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['figure.figsize'] = 12, 6
plt.rcParams['font.size'] = 14
plt.rcParams['axes.unicode_minus'] = False
```

### 선 그래프 
- 어떠한 데이터 흐름에 따라 변화되는 데이터를 표현할 때 사용

```python
X = [1, 2, 3]
y = [2, 3, 1]

plt.title('선 그래프')
plt.plot(X, y)

# 각 축의 값의 종류와 동일한 것으로 설정해야 한다
plt.xticks([1, 2, 3], ['가', '나', '다'])
plt.yticks([0, 2, 4], ['하', '중', '상'])

# 각 축의 이름
plt.xlabel('x축')
plt.ylabel('y축')
plt.show()
```

### bar chart 
- 데이터들의 양을 비교하기 위해 사용

```python
X = [1, 2, 3]
y = [2, 3, 1]

plt.title('막대 그래프')
plt.bar(X, y)

# 각 축의 값의 종류와 동일한 것으로 설정해야 한다
plt.xticks([1, 2, 3], ['가', '나', '다'])
plt.yticks([0, 2, 4], ['하', '중', '상'])

# 각 축의 이름
plt.xlabel('x축')
plt.ylabel('y축')
plt.show()
```

```python
X = [1, 2, 3]
y = [2, 3, 1]

plt.title('가로 방향 막대 그래프')
plt.barh(y, X)

# 각 축의 값의 종류와 동일한 것으로 설정해야 한다
plt.yticks([1, 2, 3], ['가', '나', '다'])
plt.xticks([0, 2, 4], ['하', '중', '상'])

# 각 축의 이름
plt.xlabel('x축')
plt.ylabel('y축')
plt.show()
```

### stem plot
- 막대 그래프와 유사하지만 폭이 없는 그래프
- 이산 확률 함수나 자기 상관 관계를 묘사할 때 사용

```python
x = np.linspace(0.1, 2 * np.pi, 10)
y = np.cos(x)

plt.title('stem plot')
plt.stem(x, y, '-.')
plt.show()
```

### pie chart
- 카테고리 별 값의 상대적인 비교를 해야 할 때
- 막대 그래프는 양을 비교하는 것이고 파이 차트는 비율을 비교

```python
values = [15, 30, 40, 10]
labels = ['값', '값2', '값3', '값4']
colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']

# 조각의 튀어나옴 정도
explode = [0, 0.1, 0, 0]

plt.title('pie chart')
# 첫번째 : 값, labels : 파이 조각의 이름, colors : 파이의 색상
# explode : 파이별로 튀어나옴 정도, shadow : 그림자 표시 여부
# startangle : 원의 시작 각도(기본인 0도는 3시 방향, 시계 반대 방향)
# counterclock : 원의 방향을 시계 반대 방향으로 할 것인가, default = True
# autopct : 파이에 설정을 비율 값 양식, 설정하지 않으며 나타나지 않음
plt.pie(values, labels=labels, colors=colors, explode=explode,
       shadow=True, startangle=90, counterclock=False,
       autopct='%.2f%%')
plt.show()
```

### histogram 
- 막대 그래프와 유사해 보이지만 구간별 양을 비교하는 용도로 사용

```python
np.random.seed(0)
x = np.random.randn(1000)
plt.title('Histogram')
plt.hist(x, bins=30)
plt.show()
```

### scatter plot
- 2차원 데이터 즉, 두 개의 실수 데이터 집합의 상관관계
- 머신러닝에서는 데이터의 산점도를 그려서 분포를 파악하고 이를 통해 분류 가능 여부를 판단 할 수 있다. 중첩되는 구간이 너무 많으면 어떤 학습 알고리즘을 사용해도 성능이 매우 떨어질 수 밖에 없다

```python
# 0 ~ 1 사이의 값 100개를 표준편차를 기반으로 값을 조정
np.random.seed(0)
X = np.random.normal(0, 1, 100)
y = np.random.normal(0, 1, 100)
plt.title('Scatter Plot')
plt.scatter(X, y)
plt.show()
```
